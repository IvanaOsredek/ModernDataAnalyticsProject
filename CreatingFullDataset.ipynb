Cluster the data per 10 minutes

```python
#First convert result_timestamp to datetime object
noise_df['result_timestamp'] = pd.to_datetime(noise_df['result_timestamp'],format="%d/%m/%Y %H:%M:%S.%f")

#Rename 'object_id' column
noise_df.rename(columns={'#object_id':'object_id'},inplace=True)

#Cluster data per 10 minutes
noise_clustered_df = noise_df.groupby(['object_id',pd.Grouper(key='result_timestamp',freq='10T')]).mean().reset_index()

noise_clustered_df
```

Add extra variables to the dataset

```python
#Create new variables with the date and time
noise_clustered_df['date'] = noise_clustered_df['result_timestamp'].dt.date
noise_clustered_df['time'] = noise_clustered_df['result_timestamp'].dt.time

#Add the hour
noise_clustered_df['hour'] = noise_clustered_df['result_timestamp'].dt.hour

#Add the day of the week
noise_clustered_df['weekday'] = noise_clustered_df['result_timestamp'].dt.day_name()

#Add the month
noise_clustered_df['month'] = noise_clustered_df['result_timestamp'].dt.month

#Add whether it's the weekend or not
noise_clustered_df['weekend'] = noise_clustered_df['weekday'].isin(['Saturday','Sunday'])

#Add the day of the week + the time
noise_clustered_df['day_time'] = noise_clustered_df['weekday'].astype(str) + " " + noise_clustered_df['time'].astype(str)

#Add the day of the week + the hour
noise_clustered_df['day_hour'] = noise_clustered_df['weekday'].astype(str) + " " + noise_clustered_df['hour'].astype(str)

#Add whether it's a holiday
#Official KU Leuven holidays: 2022-01-01 - 2022-01-13; 2022-02-02; 2022-02-06 - 2022-02-13; 2022-04-02 - 2022-04-18;
#2022-05-01; 2022-05-26; 2022-05-28 - 2022-06-12; 2022-07-03 - 2022-09-25; 2022-11-01 - 2022-11-02;
#2022-11-11; 2022-12-24 - 2022-12-31
holidays = pd.concat([pd.DataFrame(pd.date_range(start='2022-01-01',end='2022-01-13')),
                      pd.DataFrame(pd.date_range(start='2022-02-02',end='2022-02-02')),
                      pd.DataFrame(pd.date_range(start='2022-02-06',end='2022-02-13')),
                      pd.DataFrame(pd.date_range(start='2022-04-02',end='2022-04-18')),
                      pd.DataFrame(pd.date_range(start='2022-05-01',end='2022-05-01')),
                      pd.DataFrame(pd.date_range(start='2022-05-26',end='2022-05-26')),
                      pd.DataFrame(pd.date_range(start='2022-05-28',end='2022-06-12')),
                      pd.DataFrame(pd.date_range(start='2022-07-03',end='2022-09-25')),
                      pd.DataFrame(pd.date_range(start='2022-11-01',end='2022-11-02')),
                      pd.DataFrame(pd.date_range(start='2022-11-11',end='2022-11-11')),
                      pd.DataFrame(pd.date_range(start='2022-12-24',end='2022-12-31'))])
holidays.columns = ['dates']
noise_clustered_df['holiday'] = noise_clustered_df['date'].astype(str).isin(holidays['dates'].astype(str))

#Add whether it's an exam period
#Official KU Leuven exam periods: 2022-01-14 - 2022-02-05; 2022-06-13 - 2022-07-02; 2022-08-22 - 2022-09-10
exam_periods = pd.concat([pd.DataFrame(pd.date_range(start='2022-01-14',end='2022-02-05')),
                         pd.DataFrame(pd.date_range(start='2022-06-13',end='2022-07-02')),
                         pd.DataFrame(pd.date_range(start='2022-08-22',end='2022-09-10'))])
exam_periods.columns = ['dates']
noise_clustered_df['exam_period'] = noise_clustered_df['date'].astype(str).isin(exam_periods['dates'].astype(str))

noise_clustered_df
```
